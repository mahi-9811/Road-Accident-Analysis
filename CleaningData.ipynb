```python
import pandas as pd
import numpy as np
import requests  # For API data fetching (if needed)
```

```python
df = pd.read_csv("road accident.csv")
df.head()  # Show first 5 rows
```

```python
df.info()  # Check column types
df.describe()  # Check numerical stats
```

```python
print(df.isnull().sum())  # Count missing values in each column
```

```python
print(f"Duplicate rows: {df.duplicated().sum()}")  # Count duplicate rows

# Remove duplicates
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")  # Normalize column names
print(df.columns)
```

```python
cleaned_file_path = "cleaned_road_accident.csv"
df.to_csv(cleaned_file_path, index=False)
print(f"Cleaned dataset saved at {cleaned_file_path}")
```
